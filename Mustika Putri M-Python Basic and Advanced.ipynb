Contoh:

angka yang genap pada list tersebut adalah 2 dan 4, sehingga keluaran program 6


```python
list_angka=[1,2,3,4,5]
result=0
for i in list_angka:
    if (i % 2)==0:
        result+=i
print ("jumlah bilangan genap",result)
```

```python
# Fungsi Jumlah angka genap dalam list
def sum_genap(list_angka):
    jumlah_genap=0
    for angka in list_angka:
        if angka%2==0:
            jumlah_genap+=angka
    return jumlah_genap
list_angka=[1,2,3,4,5,100]
hasil = sum_genap(list_angka)
```

```python
print(f"jumlah angka genap adalah: {hasil}")
```

## Nomor 2

2.	Buat program python untuk mengecek apakah suatu list berisi angka-angka terurut(terurut naik ataupun turun) atau tidak
Contoh :
list_angka = [1,3,5,7]
karena angkanya berurut maka program python mengeluarkan boolean True


```python
def cek_urutan(angka):
    terurut_naik=True
    terurut_turun=True
    
    for i in range(len(angka)-1):
        if angka[i]>angka[i+1]:
            terurut_naik=False
        elif angka [i]<angka[i+1]:
            terurut_turun=False
    return terurut_naik or terurut_turun

angka = [1,3,5,7]
hasil=cek_urutan(angka)
print(hasil)
```

### Nomor 2

Penulisan nama kolom menggunakan spasi tidak sesuai dengan kaidah penamaan kolom yang baik. Buat program untuk mengubah nama kolom sesuai dengan kaidah penulisan camel_case

```python
def to_camel_case(nama_kolom):
    nama = nama_kolom.split()
    camel_case_name = nama[0].lower()  # Mengonversi kata pertama ke huruf kecil
    for nama in nama[1:]:
        camel_case_name += "_" + nama.lower()  # Menambahkan kata berikutnya dengan underscore dan huruf kecil
    return camel_case_name

def convert_columns_to_camel_case(df):
    new_columns = []
    for nama_kolom in df.columns:
        new_columns.append(to_camel_case(nama_kolom))
    df.columns = new_columns
    return df

# Contoh penggunaan

print("DataFrame sebelum konversi:")
print(df.info())

df = convert_columns_to_camel_case(df)

print("\nDataFrame setelah konversi:")
print(df.info())
```

### Nomor 1

Buatlah visualisasi untuk menunjukkan grossprofit setiap quarternya

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.dates as mdates
```

```python
# karena data tahunan maka lebih cocok menggunakan line
# Draw the line chart
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(16, 6))
ax.plot(df.date, df.gross_profit, label='Gross Profit per Quarter',marker='o')
ax.legend()

# increase the major tick frequency (8 ticks in this example)
start,end = plt.gca().set_xlim(pd.Timestamp('2017-03-31'), pd.Timestamp('2023-03-31'))  # Contoh batas baru
xticks= np.linspace(start,end,25)
ax.set_xticks(xticks)
ax.set_xticklabels(xticks)
# Atur ulang batas sumbu x

# set date format
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m/%y'))

# use matplotlib's auto date styling
fig.autofmt_xdate()

# Tambahkan judul
plt.title('Perkembangan Gross Profit Perusahaan Spotify per Quarter, 31-03-2017 - 31-03-2023')
plt.xlabel('Quarter')
plt.ylabel('gross_profit')
```
